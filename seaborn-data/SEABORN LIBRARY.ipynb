KDE= KERNEL DENSITY ESTIMATION

   A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. 
    
    KDE represents the data using a continuous probability density curve in one or
    more dimensions.

By default kde is True.,that is why we get these lines inside graph ,if we make it as false ,it will disappear.

```python
sns.distplot(tipdx["total_bill"],bins=50,kde=False)  # the mid line disappeared.
```

# NUMERICAL PLOTS

In numerical plot we need both "x" and "y" as numeric data. there are 3 different numeric plots .

1-- scatter plot

2-- rel plot (relational plot)

3-- line plot

```python
mpgds=sns.load_dataset("mpg")
print(mpgds)
```

1----SCATTER PLOT

```python
sns.scatterplot(x="mpg",y="weight",data=mpgds)    # with increase in mpg ,weight decreased.
```

```python
sns.scatterplot(x="horsepower",y="mpg",data=mpgds)    #increase in horsepower,mpg decreased.
```

2---- LINEPLOT

```python
sns.lineplot("weight","mpg",data=mpgds)
```

```python
sns.lineplot("model_year","mpg",data=mpgds)
```

3-- REL PLOT

```python
sns.relplot("weight","mpg",data=mpgds)
```

```python
# NOTE-- we cannot use boxplot in numerical plot. we can only use it in categorical data.
```
