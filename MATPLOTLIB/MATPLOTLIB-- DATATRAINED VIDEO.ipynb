# MATPLOTLIB-- DATATRAINED VIDEO

Artist layer

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure

fig = Figure()
canvas = FigureCanvas(fig)

import numpy as np
```

```python
x=np.random.randn(10000)
```

```python
ax=fig.add_subplot(111)#create grid with 1 row 1 column and uses the first cell for loaction of axes
```

```python
ax.hist(x,100)  #generate a histogram of 10000 numbers
```

```python
ax.set_title("Normal distribution with $\mu=0 , \sigma=1$")
```

```python
fig.savefig('matplotlib_histogram.png')
```

#### SCRIPTED LAYER -- Uses matplotlib with pyplot,a scripted layer is lighter than artist layer.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.randn(10000)
plt.hist(x,100) 
plt.title("normal distribution with $\mu=0 , \sigma=1$")
plt.savefig('matplotlib_histogram.png')
plt.show()
```
